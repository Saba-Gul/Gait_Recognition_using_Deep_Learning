```python
def generate_gait_cycle_list(folder_path, csv_file):
    for subject in csv_file:
        images_list = np.ndarray((subject[3] - subject[2] + 1, 128, 88))
    
        subject_dir = str(subject[0])
        if len(subject_dir) != 7:
            subject_dir = '0' * (7 - len(subject_dir)) + subject_dir
                
        for ii, image_name in enumerate(range(subject[2], subject[3]+1)):        
            image_name = str(image_name)
            if len(image_name) == 2:
                image_name = '0' + image_name
            elif len(image_name) == 1:
                image_name = '00' + image_name
            images_list[ii, :, :] = imread(os.path.join(folder_path, subject_dir, '00000{}.png'.format(image_name)))
            
            if ii == 0:
                begin_image = image_name
                
        yield (images_list, subject_dir, begin_image, image_name)
```

```python
def generate_save_path(folder_path, subject_dir, image_name, idx, degree):
    start = str(int(image_name) + idx)
    end = str(int(image_name) + idx + 1)
    
    if len(start) ==2:
        start = '0' + start
    elif len(end) == 2:
        end = '0' + end
    save_path = os.path.join(folder_path, subject_dir, '{}_{}_{}'.format(start, end, str(degree)))
                             
    return save_path
```

```python
Or_folder_path = 'OULP-C1V2_Pack/OULP-C1V2_NormalizedSilhouette(88x128)/Seq01'
# subject_list = os.listdir(Or_folder_path) 
# subject_list = subject_list[1:]
save_folder = 'OULP-C1V2_Pack/OULP-deference-(128x88)/Seq01'
image_list_tuple = generate_gait_cycle_list(Or_folder_path, IDList_A_55_probe)

for ii in range(len(IDList_A_55_probe)):
    next_input = next(image_list_tuple)
    subject_dir = str(next_input[1])
    begin_name = next_input[2]
    image_list = next_input[0]
    
#     os.makedirs(os.path.join(save_folder, subject_dir))
    
    for time in range(len(image_list)-1):
        error = image_list[time + 1] - image_list[time]
        
        save_path = generate_save_path(save_folder, subject_dir, begin_name, time, 85)
        
        if os.path.isdir(os.path.join(save_folder, subject_dir)):
            imsave('{}.png'.format(save_path), error)
```

# Rename

```python
path = 'OULP-C1V2_Pack/OULP-deference-(224x224)/Seq01'
lists = os.listdir(path)
lists = lists[1:]
for subject in lists:
    subject_path = os.path.join(path, subject)
    images = os.listdir(subject_path)
    for image in images:
        image_path = os.path.join(subject_path, image)
        length_image = len(image)
        
        if length_image == 15:
            new_name = image[1:] 
        elif length_image == 12:
            new_name = '0' + image
        elif length_image == 13 and image[0] != '0':
            new_name = '0' + image
        else:
            new_name = image
        
        os.renames(image_path, os.path.join(subject_path, new_name))
    
```
